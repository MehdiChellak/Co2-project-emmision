```python
concatenated_df  = pd.DataFrame()
for code in dataco2['Country code'].unique():
    dataWithCountry = dataco2[dataco2['Country code'] == code].sort_values('Year')
    datawithmean =dataWithCountry.fillna(value=dataco2.mean())
    concatenated_df = pd.concat([concatenated_df, datawithmean])
```
